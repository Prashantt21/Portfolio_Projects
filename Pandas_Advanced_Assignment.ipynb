## Applying functions to columns and creating new columns

We need to create another column in d3, which contains  a boolean value for each city to indicate whether it's a union territory or not.
- HINT: Chandigarh, Pondicherry and Delhi are only 3 union territories here.


```python
# write function here
def to_ut(x):
  return x in ('Chandigarh','Pondicherry','Delhi')

for x in d3:
  if to_ut(x):
    print("Union_Territory")
  else:
    print("State")

d3['to_ut'] = d3['city'].apply(to_ut)
```

```python
# print d3
d3
```

The temperatures mentioned in ‘Temperature’ column are mentioned in Celsius, we need another column which contains the same in Fahrenheit.

HINT -
- Define a function c_to_f which takes input temp in celsius and returns a value with temperature in Fahrenheit.
- To check: c_to_f(10) should return 50.


```python
# write function here
def c_to_f(celcius):
  return(celcius * 9/5) + 32
```

```python
# check function c_to_f(10)
c_to_f(10)

```

```python
# apply function c_to_f to d3 to create a column 'temp_farenhiet'
d3['temp_farenhiet'] = d3['Temperature'].apply(c_to_f)
d3
```

## Groupby

```python
# Create a dataframe using dictionary of your choice

data = {'Employee_ID': [1001,1002,1003,1004,1005,1006,1007.1008], 'Employee_Name':['Prasad Aher','Aman Sharma','Mayur Gayake','Adarsh Bhardwaj','Abbas Zargham','Shiv Patel','Sadik Sheikh'],'position': ['Marketing','Executive','Trainee','Executive','Marketing','Trainee','Trainee'],'Salary':[80000,35000,30000,35000,98000,25000,25000],'Dept': ['Sales','Development','Development','Sales','Sales','Development','Development'] }
d1 = pd.DataFrame(data)
d1
```

```python
# Use Groupby of single column with aggregate sum()
group_by = d1.groupby('position')['Salary'].sum()
group_by
```

```python
# Use Groupby of single column with aggregate count()
group_by = d1.groupby('position')['position'].count()
group_by
```

```python
# Use Groupby of single column with aggregate min() and max()
min_salary = d1.groupby('position')['Salary'].min()
min_salary
max_salary = d1.groupby('position')['Salary'].max()
max_salary
```

```python
# Use Groupby of any 2 columns with aggregate mean()
grouped_mean = d1.groupby('position')['Salary'].mean()
grouped_mean
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()
grouped_min_max = d1.groupby(['Dept', 'position'])['Salary'].agg(['min', 'max'])
grouped_min_max
```
